```python
#to remove dictionary from list of genre and get items only in list format
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
#o remove dictionary from list of genre and get items only in list format or objects we want only top 3 casts
import ast
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
        return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    try:
        obj = ast.literal_eval(obj)
        for i in obj:
            if i.get('job') == 'Director':
                return i.get('name')
        return None
    except:
        return None
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import ast

# Function to safely handle None or NaN values and iterate over lists
def safe_apply(x):
    if x:  # Check if x is not None or empty
        if isinstance(x, str):  # If it's a string, attempt to evaluate it
            try:
                x = ast.literal_eval(x)  # Convert string to list if possible
            except (ValueError, SyntaxError):
                return []  # Return empty list if conversion fails
        return [i.strip() for i in x]  # Strip spaces if it's a valid list
    else:
        return []  # Return empty list if x is None or invalid

# Apply the function to the relevant columns
movies['genres'] = movies['genres'].apply(safe_apply)
movies['keywords'] = movies['keywords'].apply(safe_apply)
movies['cast'] = movies['cast'].apply(safe_apply)
movies['crew'] = movies['crew'].apply(safe_apply)
```

```python
movies.head()
```

```python
#merging of columns overview,genres,keywords,crew,cast into tag column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():  # Iterating through each word in the text
        y.append(ps.stem(i))  # Stemming the word and adding it to the list
    return " ".join(y)  # Joining the list into a single string and returning it
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
#converting a collection of text documents itno a neumerical representation.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

**Recommendation Function**

```python
def recommend(movie):
    try:
        # Find the index of the given movie title
        movie_index = new_df[new_df['title'] == movie].index[0]

        # Get similarity scores for that movie
        distances = similarity[movie_index]

        # Sort movies based on similarity scores (excluding the first one, which is the movie itself)
        movies_list = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:6]

        # Print recommended movies
        for i in movies_list:
            print(new_df.iloc[i[0]].title)

    except IndexError:
        print("Movie not found in the dataset.")

    return
```

 **Function call to generate recommendations for "The Dark Knight Rises"**

```python
recommend('The Dark Knight Rises')
```

**Function call to generate recommendations for "Batman Begins"**

```python
recommend('Batman Begins')
```

**Function call to generate recommendations for "John Carter"**

```python
recommend('John Carter')
```
